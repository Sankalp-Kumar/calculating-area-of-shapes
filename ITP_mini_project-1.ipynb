### Problem 1:

1. Write a program to calculate area of shapes. Your program should be capable of calculating the area of a square, rectangle, triangle and a circle. The user should be presented with options to select the shape. Based on which shape is chosed by the user, the program should ask for the appropriate input and print the resulting area on the screen.

When the program is run, the screen should display something like this:

**Which shape would you like to calculate the area for? Please enter the option number-**<br>
**1. Square**<br>
**2. Rectangle**<br>
**3. Triangle**<br>
**4. Circle**<br>
**Enter Option: _**

Say the user enters the option 1.

**Please enter the length of a side: _**

If the user enters a value of 5. The output should be:

**The area of the square is 25**

This program should indicate that the input is invalid if the user enters a character instead of a number as input. For instance if the user enters a value of ‘a’ instead of 5 in the previous example the program should prompt:

**Invalid input, please enter a number: _**


```python
area = 0 
def square():
    global area           
    side = input("Please enter the value of side :")
    
    if side.isalpha():          #prints invalid input if the entered value is not a number
        print("Invalid input, please enter a number: _")
        square()          #This will call the square() function untill the given input is a number     
        
    else : 
        float_side = float(side)
        area= float_side**2
    return area
def Rectangle():
    global area
    leng = input("Please enter the length of the Rectangle :")
    wid = input("Please enter the width of the Rectangle :")
    
    if leng.isalpha() or wid.isalpha():     #prints invalid input if the entered values is not a number
        print("Invalid input, please enter a number: _")
        Rectangle()      #This will call the Rectangle() function untill the given input is a number
    else : 
        length = float(leng)
        width = float(wid)
        area = length*width
    return area
def Triangle():
    global area
    h = input("Please enter the height of the Triangle :")
    b = input("Please enter the base value of the Triangle :")
    
    if h.isalpha() or b.isalpha():         #prints invalid input if the entered values is not a number
        print("Invalid input, please enter a number: _")
        Triangle()        #This will call the Triangle() function untill the given input is a number
        
    else : 
        height = float(h)
        Base = float(b)
        area = 1/2*height*Base
    return area
def Circle():
    global area
    pi = 3.141
    radius = input("Please enter the Radius of the Circle :")
    
    if radius.isalpha():         #prints invalid input if the entered value is not a number
        print("Invalid input, please enter a number: _")
        Circle()      #This will call the Circle() function untill the given input is a number
        
    else : 
        rad = float(radius)
        area = pi*(rad**2)
    return area

ui = input(""""Which shape would you like to calculate the area for? Please enter the option number-
1. Square
2. Rectangle
3. Triangle
4. Circle
Enter Option: _ """)
print(ui)


if ui == str("1"):
    square()
    print("The area of the square is :",area)
elif ui == str("2"):
    Rectangle()
    print("The area of the Rectangle is :",area)
elif ui == str("3"):
    Triangle()
    print("The area of the Triangle is :",area)
elif ui == str("4"):
    Circle()
    print("The area of the Circle is :",area)
else :
    print("Invalid input, please enter a number between 1-4: _")
    
```
